```python
def array_properties(a):
    print("array: ", a)
    print("array shape:", a.shape)
    print("array dimension: ", a.ndim)
    print("array datatype: ", a.dtype)
    print("array size: ", a.size)
    print()
```

```python
#Q1. Create an array of shape (100, 100, 3) of unsigned int

import numpy as np

np.random.seed(12345)

a = np.random.randint(0, 256, (100, 100, 3), dtype = np.uint8)
array_properties(a)
```

```python
#Q2 Find the values that are less than 30, 50, 100, 250 

a30 = a[a<30]
print("Values less than 30:", a30)

a50 = a[a<50]
print("Values less than 50:", a50)

a100 = a[a<100]
print("Values less than 100:", a100)

a250 = a[a<250]
print("Values less than 100:", a250)
```

```python
ia30_ = a<30
ia30_ = ia30_ == True 
print(ia30_)
print("Values less than 30:", a[ia30_])

print()

ia50_ = a<50
ia50_ = ia50_ == True 
print(ia50_)
print("Values less than 50:", a[ia50_])

print()

ia100_ = a<100
ia100_ = ia100_ == True 
print(ia100_)
print("Values less than 100:", a[ia100_])

print()

ia250_ = a<250
ia250_ = ia250_ == True 
print(ia250_)
print("Values less than 250:", a[ia250_])
```

```python
#Q3 Split the array into 4 parts vertically and horizontally

a1 = np.hsplit(a, 4)
print(a1)
print(len(a1))
print(type(a1))
```

```python
a1 = np.vsplit(a, 4)
print(a1)
print(len(a1))
print(type(a1))
```

```python
#Q4 Copy array from question 1 and change the values that are less than 110 to 0 and others to 255

import numpy as np

np.random.seed(12345)

a = np.random.randint(0, 256, (100, 100, 3), dtype = np.uint8)
print("Array properties a: ")
array_properties(a)

b = a.copy()
print("Array properties b: ")
array_properties(b)

```

```python
b[b<110] = 0
b[b>=110] = 255
```
