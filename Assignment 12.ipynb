```python
import numpy as np

def create_canvas(height=200, width=200, bgColor=0): 
    # create the image (canvas)
    canvas = np.zeros((height, width, 3), dtype='uint8')
    if bgColor != 0:
        canvas[:] = bgColor
    return canvas

def show_in_matplotlib(img, figsize=(6,8)):
    imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.imshow(imgRGB)
    plt.title("Polygon")
    plt.show()
```

```python
# Color Dictionary 

colors = {
    'Silver' : (192,192,192),
    'Olive' : (128,128,0),
    'Midnight' : (220,20,60),
    'Tomato' : (71,99,255),
    'Lime' : (0,255,0),
    'Sienna' : (160,82,45),

}
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


img = create_canvas(height=400, width=400, bgColor=colors['Silver'])

#Triangle
pts = np.array([[40,350],[40,250],[150,350]], np.int32)  
pts = pts.reshape((-1, 1, 2)) 
cv.polylines(img, [pts], True, colors['Olive'], 12)

#Hexagon 
pts = np.array([[120,48],[60,48],[30,100],[60,152],[120,152],[150,100]], np.int32)  #3 by 2
pts = pts.reshape((-1, 1, 2)) 
cv.polylines(img, [pts], True, colors['Midnight'], 4)

#Decagon
pts = np.array([[335,234],[285,234],[245,263],[230,310],[245,357],[285,386],[335,386],
                [375,357],[390,310], [375,263]], np.int32)  #3 by 2
pts = pts.reshape((-1, 1, 2)) 
# draw triangle
cv.polylines(img, [pts], True, colors['Tomato'], 5)

# 7 Points 
pts = np.array([[400,320], [470,440], [530,280], [660,300], [710,110], [570,180],
               [436,100]])
pts = np.int32((pts * 0.5))
pts = pts.reshape((-1, 1, 2))
# draw triangle
cv.polylines(img, [pts], False, colors['Lime'], 9)

show_in_matplotlib(img)
```
